```python
def loan_status_is_great(loan_status):
  if (loan_status == 'Current') or (loan_status == 'Fully Paid'):
    return 1
  else: 
    return 0 
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loan_status_is_great)
```
